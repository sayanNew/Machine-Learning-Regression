```python
#1. What is Simple Linear Regression?

'''
Simple Linear Regression is a statistical and machine learning method used to predict a continuous value using one independent variable.
It models the relationship between an independent variable (X) and a dependent variable (Y) using a straight line.
The equation of the line is Y = a + bX, where a is the intercept and b is the slope.
It is commonly used to predict values such as salary based on experience or price based on demand.
'''
```

```python
#2. What are the key assumptions of Simple Linear Regression?

'''
The main assumptions of Simple Linear Regression are:
Linearity – There is a linear relationship between the independent variable (X) and the dependent variable (Y).
Independence – The observations or errors are independent of each other.
Homoscedasticity – The variance of the errors is constant for all values of X.
Normality of Errors – The residuals (errors) are normally distributed.
These assumptions ensure that the regression model produces reliable and accurate predictions.
'''
```

```python
#3. What does the coefficient m represent in the equation Y=mX+c?

'''
In the equation Y = mX + c used in Simple Linear Regression, the coefficient m represents the slope of the line.
It shows how much the dependent variable (Y) changes when the independent variable (X) increases by one unit.
If m is positive, Y increases as X increases; if m is negative, Y decreases as X increases.
'''
```

```python
#4. What does the intercept c represent in the equation Y=mX+c?

'''
In the equation Y = mX + c used in Simple Linear Regression, the intercept (c) represents the value of Y when X = 0.
It is the point where the regression line intersects the Y-axis on the graph.
In simple terms, it shows the starting or baseline value of Y before X has any effect.
'''
```

```python
#5. How do we calculate the slope m in Simple Linear Regression?

'''
In Simple Linear Regression, the slope (m) is calculated using the formula:

     m = (n∑XY−(∑X)(∑Y))/(n∑pow(X,2) - pow(∑X, 2))


Where:

n = number of data points
X = independent variable values
Y = dependent variable values
ΣXY = sum of the product of X and Y
ΣX = sum of X values
ΣY = sum of Y values
This formula finds the best-fit line that shows how much Y changes for each unit increase in X.
'''
```

```python
#6.  What is the purpose of the least squares method in Simple Linear Regression?

'''
In Simple Linear Regression, the Least Squares Method is used to find the best-fit straight line for the data.
Its purpose is to minimize the sum of the squared differences between the actual values (Y) and the predicted values (Ŷ).
By minimizing these squared errors, the method determines the optimal slope (m) and intercept (c) for the regression line, making the model’s predictions as accurate as possible.
'''
```

```python
#7. How is the coefficient of determination (R²) interpreted in Simple Linear Regression?

'''
In Simple Linear Regression, the coefficient of determination (R²) measures how well the regression model explains the variation in the dependent variable.
R² represents the proportion of the variance in Y that is explained by the independent variable X.
Its value ranges from 0 to 1.
R² close to 1 → the model fits the data well.
R² close to 0 → the model explains very little of the variation.
'''
```

```python
#8. What is Multiple Linear Regression?

'''
Multiple Linear Regression is a machine learning technique used to predict a dependent variable using two or more independent variables.
It models the relationship between one output variable and several input variables using a linear equation.
The general form is Y = a + b₁X₁ + b₂X₂ + ... + bₙXₙ.
It is commonly used in problems like predicting house prices using size, location, and number of rooms.
'''
```

```python
#9. What is the main difference between Simple and Multiple Linear Regression?

'''
The main difference between Simple Linear Regression and Multiple Linear Regression is the number of independent variables used.
Simple Linear Regression uses one independent variable (X) to predict the dependent variable (Y).
Multiple Linear Regression uses two or more independent variables (X₁, X₂, …) to predict the dependent variable.
In short, simple regression has one predictor, while multiple regression has multiple predictors.
'''
```

```python
#10. What are the key assumptions of Multiple Linear Regression?

'''
The key assumptions of Multiple Linear Regression are:
Linearity – There is a linear relationship between the dependent variable and the independent variables.
Independence of errors – The residuals (errors) are independent of each other.
Homoscedasticity – The variance of the errors is constant across all levels of the independent variables.
Normality of errors – The residuals are normally distributed.
No multicollinearity – Independent variables should not be highly correlated with each other.
'''
```

```python
#11. What is heteroscedasticity, and how does it affect the results of a Multiple Linear Regression model?

'''
In Multiple Linear Regression, heteroscedasticity occurs when the variance of the error terms (residuals) is not constant across all values of the independent variables.
This violates one of the key regression assumptions. As a result, the estimated coefficients may still be unbiased, but the standard errors become unreliable, which can lead to incorrect hypothesis tests and confidence intervals.
'''
```

```python
#12. How can you improve a Multiple Linear Regression model with high multicollinearity?

'''
In Multiple Linear Regression, multicollinearity occurs when independent variables are highly correlated. You can improve the model using these methods:
Remove highly correlated variables to reduce redundancy.
Combine correlated variables (e.g., create a new feature).
Use regularization techniques like Ridge Regression or Lasso Regression.
Collect more data to reduce the effect of correlation among variables.
These steps help make the regression coefficients more stable and reliable.
'''
```

```python
#13. What are some common techniques for transforming categorical variables for use in regression models?

'''
In regression models such as Multiple Linear Regression, categorical variables must be converted into numerical form before training the model. Common techniques include:
One-Hot Encoding – Creates a separate binary column (0 or 1) for each category.
Label Encoding – Assigns a unique numeric value to each category.
Dummy Variable Encoding – Similar to one-hot encoding but removes one category to avoid multicollinearity.
Ordinal Encoding – Used when categories have a natural order (e.g., low, medium, high).
These methods allow regression algorithms to process categorical data effectively.
'''
```

```python
#14. What is the role of interaction terms in Multiple Linear Regression?

'''
In Multiple Linear Regression, interaction terms are used to examine whether the effect of one independent variable on the dependent variable changes depending on another variable.
They are created by multiplying two independent variables (e.g., X1 * X2)
Interaction terms help the model capture combined or joint effects between variables, improving the model’s ability to represent complex relationships.
'''
```

```python
#15.How can the interpretation of intercept differ between Simple and Multiple Linear Regression?

'''
In Simple Linear Regression, the intercept represents the value of the dependent variable when the single independent variable (X) is 0.
In Multiple Linear Regression, the intercept represents the value of the dependent variable when all independent variables (X₁, X₂, …, Xₙ) are 0.
So, in simple regression it depends on one variable, while in multiple regression it represents the baseline when all predictors are zero.
'''
```

```python
#16.What is the significance of the slope in regression analysis, and how does it affect predictions?

'''
In regression analysis, the slope indicates how much the dependent variable changes when the independent variable increases by one unit.
In Simple Linear Regression, the slope shows the strength and direction of the relationship between X and Y.
A positive slope means Y increases as X increases, while a negative slope means Y decreases as X increases.
Thus, the slope directly affects predictions by determining how the predicted value of Y changes with X.
'''
```

```python
#17.How does the intercept in a regression model provide context for the relationship between variables?

'''
In a regression model, the intercept represents the expected value of the dependent variable when all independent variables are zero.
In Simple Linear Regression, it shows the baseline starting value of Y before X has any effect.
It provides context by giving a reference point for the regression line, helping interpret how the dependent variable changes as the independent variables increase.
'''
```

```python
#18. What are the limitations of using R² as a sole measure of model performance?

'''
The coefficient of determination (R²) is useful, but relying on it alone has several limitations in regression models like Multiple Linear Regression.
Does not detect overfitting – A model may have a high R² but still perform poorly on new data.
Always increases with more variables – Adding more predictors can increase R² even if they are not meaningful.
Does not indicate model correctness – A high R² does not guarantee that the model assumptions (linearity, independence, etc.) are satisfied.
Does not measure prediction accuracy alone – Other metrics like RMSE or MAE are also needed.
So, R² should be used together with other evaluation metrics for better model assessment.
'''
```

```python
#19. How would you interpret a large standard error for a regression coefficient?

'''
A large standard error for a regression coefficient indicates that the estimate of the coefficient is not very precise.
It means the coefficient may vary a lot if the sample data changes, suggesting uncertainty in the estimate.
In models like Multiple Linear Regression, this often implies that the predictor may not have a strong or reliable effect on the dependent variable.
'''
```

```python
#20. How can heteroscedasticity be identified in residual plots, and why is it important to address it?

'''
In Multiple Linear Regression, heteroscedasticity can be identified using a residual plot (residuals vs. predicted values).
If the residuals are randomly scattered with equal spread, the variance is constant.
But if the residuals form a pattern such as a funnel or cone shape (spread increasing or decreasing), it indicates heteroscedasticity.
It is important to address heteroscedasticity because it can make standard errors unreliable, leading to incorrect hypothesis tests and confidence intervals, which affects the reliability of the regression results.
'''
```

```python
#21. What does it mean if a Multiple Linear Regression model has a high R² but low adjusted R²?

'''
In Multiple Linear Regression, if the model has a high R² but a low adjusted R², it usually means that some independent variables are not actually useful.
R² increases whenever more variables are added to the model, even if they do not improve prediction.
However, adjusted R² penalizes unnecessary variables.
So this situation suggests that the model may contain irrelevant predictors or possible overfitting, and some variables should be removed.
'''
```

```python
#22. Why is it important to scale variables in Multiple Linear Regression?

'''
In Multiple Linear Regression, scaling variables ensures that all independent variables are on a similar range or scale.
Without scaling, variables with larger values may dominate the model and affect coefficient interpretation.
Scaling also improves numerical stability and model performance, especially when using techniques like Ridge Regression or Lasso Regression.
'''
```

```python
#23. What is polynomial regression?

'''
Polynomial Regression is a type of regression used when the relationship between the independent variable and the dependent variable is non-linear.
Instead of fitting a straight line, it fits a curved line (polynomial curve) to the data by including powers of the variable (like X^2, X^3).
This allows the model to capture more complex patterns in the data.
'''
```

```python
#24. How does polynomial regression differ from linear regression?

'''
The difference between Polynomial Regression and Linear Regression is in the relationship they model.
Linear Regression fits a straight line to represent the relationship between variables.
Polynomial Regression fits a curved line by including powers of the independent variable (like 
𝑋^2, 𝑋^3).
Polynomial regression is used when the relationship between variables is non-linear.
'''
```

```python
#25. When is polynomial regression used?

'''
Polynomial Regression is used when the relationship between the independent variable and the dependent variable is non-linear.
It is applied when a straight line (as in Linear Regression) cannot accurately represent the data pattern.
Polynomial regression helps model curved relationships by including higher-order terms (like 
𝑋^2, 𝑋^3).
'''
```

```python
#26. What is the general equation for polynomial regression?

'''
The general equation for Polynomial Regression is:
 Y = a + b1X + b2X^2 + b3X^3 + bnX^n
Where:

Y = dependent variable
X = independent variable
a = intercept
b₁, b₂, …, bₙ = coefficients of the polynomial terms
n = degree of the polynomial
This equation allows the model to capture curved (non-linear) relationships between variables.
'''
```

```python
#27. Can polynomial regression be applied to multiple variables?

'''
Yes, Polynomial Regression can be applied to multiple variables.
In this case, the model includes polynomial terms of each variable and their interactions (e.g., X1^2, X2^2, X1, X2)..
This allows the model to capture complex, non-linear relationships between several independent variables and the dependent variable.
'''
```

```python
#28. What are the limitations of polynomial regression?

'''
Some limitations of Polynomial Regression are:
Overfitting – High-degree polynomials may fit the training data too closely and perform poorly on new data.
Complexity – As the polynomial degree increases, the model becomes harder to interpret.
Sensitive to outliers – Extreme values can strongly affect the curve.
Poor extrapolation – Predictions outside the data range can become highly unrealistic.
'''
```

```python
#29. What methods can be used to evaluate model fit when selecting the degree of a polynomial?

'''
When selecting the degree of Polynomial Regression, several methods can be used to evaluate model fit:
R² and Adjusted R² – Measure how well the model explains the variance in the data.
Cross-Validation – Tests model performance on different subsets of the data.
Error Metrics – Such as Mean Squared Error (MSE) or RMSE to measure prediction error.
Residual Analysis – Checking residual plots to see whether the model fits the data properly.
These methods help choose a polynomial degree that balances good fit and avoids overfitting.
'''
```

```python
#30. Why is visualization important in polynomial regression?

'''
Visualization is important in Polynomial Regression because it helps see the non-linear relationship between the independent and dependent variables.
By plotting the data points and the fitted curve, we can check how well the polynomial curve fits the data.
Visualization also helps detect overfitting or underfitting and assists in choosing the appropriate polynomial degree.
'''
```
